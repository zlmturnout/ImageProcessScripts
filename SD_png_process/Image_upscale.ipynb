```python
import cv2,os,sys
from PIL import Image
```

```python

from cv2 import dnn_superres

# Create an SR object - only function that differs from c++ code
sr = dnn_superres.DnnSuperResImpl_create()

# Read image
image = cv2.imread(os.path.abspath('../img/0392.jpg'))

# Read the desired model
model_file = "../models/EDSR_x4.pb"
sr.readModel(os.path.abspath(model_file))

# Set the desired model and scale to get correct pre- and post-processing
sr.setModel("edsr", 1)

# Upscale the image
result = sr.upsample(image)

# Save the image
cv2.imwrite("../img/upscaled_0392.png", result)
```

```python
# 打开图像
img_file=os.path.abspath('../img/cindy.webp')
image = Image.open(img_file)

# 获取原始尺寸和DPI
width, height = image.size
original_dpi = image.info.get('dpi', (72, 72))  # 默认DPI为72 DPI
print(f"Original DPI: {original_dpi}")

# 目标DPI
target_dpi = (300, 300)

# 计算新尺寸
new_width = int(width * (target_dpi[0] / original_dpi[0]))
new_height = int(height * (target_dpi[1] / original_dpi[1]))

# 调整图像大小并保存
resized_image = image.resize((new_width, new_height), Image.ANTIALIAS)
resized_image.save("../img/upscaled_cindy.png", dpi=target_dpi)  # 设置DPI信息为300 DPI
```
