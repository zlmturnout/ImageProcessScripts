**Add nodes**

```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.DiGraph()
G.add_node('Alice')
G.add_node('Bob')
G.add_node("Chunk")
```

```python
list(G.nodes())
```

**Add edge**

```python
G.add_edge('Alice','Bob')
G.add_edge('Alice','Chunk')
G.add_edge('Bob','Alice')
G.add_edge('Bob','Chunk')
list(G.edges())
```

Now draw plot

```python
nx.draw_circular(G,node_color="tab:red",
                 node_size=2000,
                 with_labels=True)
```
